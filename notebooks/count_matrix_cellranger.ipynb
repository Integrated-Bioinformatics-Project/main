```python
! mkdir ./index/grch38
```

```python
! chmod ugo+rwx ./index/grch38
```

```python
! ~/STAR-2.7.9a/bin/Linux_x86_64/STAR \
    --runThreadN 5 \
    --runMode genomeGenerate \
    --genomeDir ./index/grch38 \
    --genomeFastaFiles ./GRCh38_latest_genomic.fna \
    --sjdbGTFfile ./GRCh38_latest_genomic.gff \
    --sjdbGTFtagExonParentTranscript Parent \
    --sjdbOverhang 99
```
